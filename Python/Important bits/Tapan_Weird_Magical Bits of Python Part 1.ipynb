Python is a very flexible programming language.

We can do tons of stuff in Python that would otherwise result in errors if we had tried it with other programming languages.

Python is an OOP language. Everything in Python is an object. And Python being flexible takes it too literally. 

You can assign functions to a variable and then execute them, pass functions as parameters. You can even declare functions inside functions and have a function return a function !

# Copying and assigning function to a variable

By assigning fn to a variable, Python creates a copy of that fn and even if you delete the fn, the variable will still work.

# Creating functions inside a function
You can even create multiple functions inside a function and execute them !

```
def main_func(name = 'Tapan'):
  print("Good evening !!")

  def greetings():
    print("Hello Tapan !")
  
  def another_greetings():
    print("Hello stranger !")

  greetings()
  another_greetings()
```

```
main_func()
```

#Returning functions

Suppose you create a bunch of functions inside a function. Note that you can only access those functions inside main_fn.

But what if you want to use those function outside the main_fn? Simple, you return a function from a function

```
def main_func(name = 'Tapan'):
  print("Good evening !!")

  def greetings():
    print("Hello Tapan !")
  
  def another_greetings():
    print("Hello stranger !")

  if name == 'Tapan':
    return greetings
  
  else:
    return another_greetings
```

```
some_func = main_func()
```

```
some_func
```

```
some_func()
```

```
some_func_2 = main_func("Harry")
```

```
some_func_2()
```
