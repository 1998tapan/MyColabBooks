#*args 

*args are used to accept arbitary number of params in fn in the form of tuples

#**kwargs 

**kwargs are used to accept arbitary number of inputs in fn in the form of dictionaries.

Note that args and kwargs are defined by * and ** respectively. We can use any names to use them.



```
def myfunc(*args):
  print(args)
  return sum(args)
```

```
myfunc(10,20)
```

```
myfunc(23,15,65,76)
```

```
def myfunc2(**kwargs):
  print(kwargs)
  print(f"I will have {kwargs['food']} for dinner")
```

```
myfunc2(food = 'eggs',fruit = 'apple')
```

```
def myfunc3(*args,**kwargs):
  print(args)
  print(kwargs)
  print(f"I will have {args[0]} {kwargs['food']} for dinner")
```

```
myfunc3(10,2,34,fruit='orange',vegetable='cauliflower',food='yogurts')
```
