#range()

range(stop) - from 0 to stop-1

range(start,stop) - from start to stop-1

range(start,stop,step) - from start to stop-1 in step size.

range() is a generator...meaning it doesn't store values, it generates value when executed.

Hence just calling range(2,20) would just return range object. To store the numbers, use list.

nums = list(range(2,30))

```
for i in range(10):
  print(i)
```

```
for i in range(1,10):
  print(i)
```

```
for i in range(1,10,2):
  print(i)
```

# enumerate()

used to iterate over something via loops.

enumerate() is useful because it provides index as well as value, hence providing an alternative over using a counter var to keep track of index.

Like range(), enumerate is also a generator.


```
enumerate(li)
```

```
li1 = list(enumerate('abcde'))
print(li1)
```

```
#tuple packing
li = ['a','b','c','d','e']

for items in enumerate(li):
  print(items)
```

```
#tuple unpacking

for index,letter in enumerate(li):
  print(f'At index {index} is letter {letter}')
```

```
#even dict.items() uses tuple packing

d = {"one":1,"two":2,"three":3,"four":4}

for items in d.items():
  print(items)
```

```
#tuple unnpacking
for keys,values in d.items():
  print(f'At key {keys} is value {values}')
```

#zip()

zip() is used to combine 2 or more lists.

zip() is also a generator, hence to store results of zip() we need to save it to var.

```
l1=[1,2,3,4,5]
l2=['a','b','c','d','e']
```

```
print(zip(l1,l2))
```

```
for num,letter in zip(l1,l2):
  print(f'first item is {num} and second item is {letter}')
```

```
l3=list(zip(l1,l2))
```

```
#tuple packing
l3
```
