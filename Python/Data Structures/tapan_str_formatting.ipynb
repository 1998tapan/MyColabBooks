```
""" Other than concat (i.e +),string can be formatted via 2 more ways
    1) format()
    2) f-strings"""
```

```
# 1) format()
```

```
random_str="hello"
```

```
print(random_str)
```

```
print("This string is {}".format("formatted"))
```

```
print("The {} {} {}".format("fox",'quick','brown'))
```

```
print("The {2} {1} {0}".format("fox","brown","quick"))
```

```
print("The {q} {b} {f}".format(f="fox",q="quick",b="brown"))
```

```
#float formatting
result=123.4353534
print("The result is {r}".format(r=result))
# var_name:width.precison f
print("The result is {r:1.3f}".format(r=result))
```

```
type(result)
```

```
print("The result is {r}".format(r=result:1.4f))
```

```
# 2) f-strings
""" f-strings stand for formatted string literals.
They were introduced in Python 3.6
"""
```

```
name="Harry"
age="11"
print("{n} got his letter at age {a}".format(a=age,n=name))
print(f'{name} got his letter at age {age}')
```

```
print("The result is {r:1.3f}".format(r=result))
print(f"The result is {result:1.3f}")
print(f"The result is {result:{1}.{3}f}")
```
