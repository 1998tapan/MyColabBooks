#Dictionaries

(covers advance lecture)

Dictionaries are unordered datatype that's similar to HashMap in Java.

Since dictionaries are unordered, indexing and slicing operations are not possible.

Dictionaries uses curly brackets.

The Keys of the dictionaries must be a String.

Dictionaries can have multiple datatypes as values be it a integer, float , string or even a list or a dictionary.

```
my_dict = {"Key 1":"value 1", "key 2":"value 2", "key 3":"value 3"}
```

```
my_dict
```

```
my_dict[1]
```

```
my_dict["key 2"]
```

```
prices = {"apples":100, "oranges": 78, "mango": 120}
print(prices["apples"])
print(prices["mango"])
```

```
my_dict = {"key1":123, "key2":[1,2,3], "key3":{"inside key":123}}
```

```
my_dict
```

```
my_dict["key2"]
```

```
my_dict["key2"][0]
```
