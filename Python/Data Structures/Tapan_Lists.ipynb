insert()

insert(index, value) can be used to insert objects/values at any index of the list.

Note that, insert() doesn't return anything.

```
new_list = ["apple","oranges", "pineapple"]
new_list.insert(1,"bananana")
print(new_list)
```

```
new_list.insert(0,1)
```

```
new_list
```

pop()
<br>
a pop() with no params will remove the last object of the list.
<br>
a pop(index) with index of a particular object will remove that object from the list.
<br>Note that the pop() also returns the removed object

```
#removes last object i.e seven
count_list.pop()
```

```
count_list
```

```
#removes object at index 2 i.e three
count_list.pop(2)
```

```
count_list
```

```
popped = count_list.pop(0)
print(f"{popped} has been removed from the list")
print(f"Updated list is {count_list}")
```

sort()


---


sort() can sort list alphabetically as well as numerically.


Note that sort() will only work if all elements of list are either alphabetic or numeric.

Note that sort() doesn't return anything so<br>
my_sorted_list = num_list.sort() 
<br>or<bR>
print(num_list.sort()) 
<br> will not work.

Sorting list in descending order:
new_list.sort(reverse = True)

or 

new_list.sort().reverse()



```
num_list = [1,2,3,4,5]
str_list = ["mango", "apple", "cherry"]
```

```
print(num_list.sort())
```

```
num_list.sort()
str_list.sort()
```

```
print(num_list)
print(str_list)
```

```
print("reversing")
num_list.sort(reverse= True)
str_list.sort(reverse= True)
print(num_list)
print(str_list)
```

```
mixed_list = ["cherry", 3, "apple", 1]
```

```
mixed_list.sort()
print(mixed_list)
```

reverse()

reverses the list, duh

Note: just like sort(), reverse() doesn't return anything.

```
print(mixed_list)
mixed_list.reverse()
print(mixed_list)
```

```
print(num_list)
num_list.reverse()
print(num_list)

print(str_list)
str_list.reverse()
print(str_list)
```
