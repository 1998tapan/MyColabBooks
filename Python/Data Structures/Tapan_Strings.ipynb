```
s.title()
```

#upper() and lower()

```
s.upper()
```

```
s.lower()
```

#count()
returns total occurences of the param in string

```
s.count('o')
```

find()
returns the first index of the param in string

```
s.find('o')
```

#center()
centers your string with a bunch of user-inputted char

```
"Tapan Dabrai".center(20,'-')
```

expandtabs()

```
"Hero\tki\tkya\tbaat\thai".expandtabs()
```

#isalnum()
checks if all characters are alphanumeric

#isalpha()
checks if all characters are alphabetic

#istitle()
checks if title string

#islower()
checks if all characters are lower case

#isupper()
checks if all characters are upper case

#isspace()
checks if all characters are space

```
print(s.isalnum())
print(s.isalpha())
print(s.istitle())
print(s.islower())
print(s.isupper())
print(s.isspace())
```

#startswith()
checks if string starts with a certain char
It's similar to s[0] == 'c'

#endswith()
checks if string ends with a certain char
It's similar to s[-1] == 'c'

```
s.startswith('h')
```

```
s.endswith('o')
```

#split()
splits the string by every occurence of inputted char

```
s.split('o')
```

#partition()

retuns a string in 3 parts:

  1)contents before first occurence of inputted char

  2)the inputted char
  
  3)contents after the first occurence of inputted char

```
s.partition('o')
```
