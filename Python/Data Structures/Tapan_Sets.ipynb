#Sets

Sets are unordered data structure which only stores unique values.

Sets won't add duplicate values but it will not give any error.

initialisation:

mySet = set()
mySet = set(myList)

methods:

```
mySet = set()
```

```
mySet.add(1)
```

# copy()

copies elements to a new set. Changes made to either set wont affect other.

But since set,list,dict are reference datatypes, changes will affect if you do this:

s = {1,2,3}

s1 = s

s1.append(4) - will affect s too since both point to same object

```
s = {1,2,3}
```

```
s1=s.copy()
```
