Tuples

Tuples are similar to Lists except the lists are mutable.

Tuples are immutable and are declared with ()

Tuples only have two methods:

  1)index()

  2)count()

Tuples are ordered , so slicing and indexing is possible.

Because tuples are immutable, you can't add new objects once it's declared and defined.

Like Lists and Dictionaries, tuples can store multiple datatypes.
