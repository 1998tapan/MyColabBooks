#Maps
With maps,one can map fn to an iterablr object. Meaning, a fn can be called repeatedly for every item in iterable

Maps are generators

They can be used instead of for loops calling methods with iterable item

```
def square(n):
  return n**2
```

```
map(square,(2,3,4,5))
```

```
list(map(square,(2,3,4,5)))
```

```
nums = [2,3,4,5,6,7,8,9]
for item in map(square,nums):
  print(item)
```

```
def splicer(word):
  if len(word) % 2 == 0:
    return 'even'
  return word[0]
```

```
words = ['Adam','DriverS','Eve']
list(map(splicer,words))
```

#Filters
Filters returns those items from iterable who when called in fn returns True. Meaning the fn must return a boolean and only those items in iterable that would return True when passed to fn.

Filters are generators

Note : function will always return boolean. Filter will return items that fn(item) = True

```
def is_even(num):
  return num % 2 == 0
```

```
nums
```

```
list(filter(is_even,nums))
```

```
for item in filter(is_even,nums):
  print(item)
```

#Lambda

Lambda are anonymous fn that are used for only one time
without needing to create a fn properly

Lambda's body is a single expr, not a block of statements.

Lambda is used for designing simple fn.

We can us if and else in lambdas similar to list comprehensions.



```
def square(n):
  return n**2
```

```
lambda n:n**2
```

```
#dont ever use this way, use map 
#This is for demo
square = lambda n:n**2
```

```
square(2)
```

```
#proper use of lambda
list(map(lambda num:num**2,nums))
```

```
list(filter(lambda num:num % 2 == 0,nums))
```

```
list(map(lambda word:word[0],words))
```

```
list(map(lambda word:word[::-1],words))
```
