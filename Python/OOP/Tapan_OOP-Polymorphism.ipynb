```
class Dog:
  def __init__(self,name):
    self.name = name

  def speak(self):
    print(self.name,"says Woof!")
```

```
class Cat:
  def __init__(self,name):
    self.name = name

  def speak(self):
    print(self.name,"says Meow")
```

```
lokie = Dog('lokie')
fredie = Cat('fredie')
```

```
def speak(pet):
    pet.speak()
```

```
speak(lokie)
```

```
speak(fredie)
```
