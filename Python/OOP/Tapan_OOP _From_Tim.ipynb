```
# Classes and Objects
```

```
class Student:
  def __init__(self, name, age, grade):
    self.name = name
    self.age = age
    self.grade = grade  # 1 -100

  def get_grade(self):
    return self.grade
```

```
class Course:
  def __init__(self, name, max_students):
    self.name = name
    self.max_students = max_students
    self.students = []

  def add_students(self, student):
    if len(self.students) <= self.max_students:
      self.students.append(student)
      return True
    return False
  
  def get_average_grade(self):
    avg = 0
    for student in self.students:
      avg += student.get_grade()
    return avg / len(self.students)
```

```
s1 = Student("Tapan", 21, 75)
s2 = Student("Kajal", 90, 80)
```

```
c1 = Course("Science", 2)
```

```
c1.add_students(s1)
c1.add_students(s2)
```

```
c1.get_average_grade()
```

# Inheritance

 - overloading and super()

```
class Pet:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  def speak(self):
    print("I dont know what I say")
  
  def show(self):
    print(f"I am {self.name} and I am {self.age} years old")
```

```
p = Pet("Gen", 12)
p.speak()
p.show()
```

```
class Dog(Pet):
  def __init__(self, name, age, breed):
    super().__init__(name, age)
    self.breed = breed
  
  def speak(self):
    print("Bow Bow")
```

```
d = Dog("Tommy", 3, "Labra")
d.speak()
d.show()
```

```
class Cat(Pet):
  def speak(self):
    print("Meow")
```

```
c = Cat("Crookshanks", 7)
c.speak()
c.show()
```

# class attributes, class methods and static methods

```
class Person:
  num_of_person = 0

  def __init__(self):
    Person.num_of_person += 1

  @staticmethod
  # doesn't need any parameter
  # don't change anything
  # final
  def msg():
    print("Hi There, I am accessible without any parameters")

  @classmethod
  # needs a class parameter
  def add_person(cls):
    cls.num_of_person += 1

  @classmethod
  def show_person(cls):
    print(cls.num_of_person)
```

```
Person.msg()
Person.show_person()
p1 = Person()
p1.show_person()
p2 = Person()
p1.show_person()
```

```
Person.show_person()
p1.show_person()
p2.show_person()
```
