```
class Dog():
  species = 'mammal'
  
  def __init__(self,name,breed,spots):
    self.name = name
    self.breed = breed
    self.spots = spots

  def display(self):
    print(f'This cute little {self.breed} is {self.name}.\nDoes he have spots? {self.spots}')

  def bark(self,num=1):
    for i in range(num):
      print("Woof!")
```

```
my_dog = Dog("Sammy","Lab",False)
my_dog.display()
my_dog.bark(2)
```

```
my_another_dog = Dog(breed="German Shephard", name="Marcus", spots=True)
my_another_dog.display()
```

```
class Circle():
  pi = 3.14

  def __init__(self,radius):
    self.radius = radius
    self.circumference = 2 * Circle.pi * self.radius

  def display(self):
    print(f"pi : {Circle.pi}")
    print(f"Radius : {self.radius}")
    print(f"Circumference : {self.circumference}")
    print(f"Area : {self.get_area()}")

  def get_area(self):
    return Circle.pi * (self.radius ** 2)
```

```
my_circle = Circle(10)
my_circle.display()
```
