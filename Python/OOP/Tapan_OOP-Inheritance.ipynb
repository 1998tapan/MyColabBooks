```
class Animal:
  def __init__(self,name,eyes,hands,legs,fins):
    self.name =  name
    self.eyes = eyes
    self.hands = hands
    self.legs = legs
    self.fins = fins

  def eat(self):
    print('Animal is eating')

  def swimming(self):
    if self.fins != 0:
      print('Animal is swimming')
  
  def run(self):
    print('Animal is running')
```

```
class Dog(Animal):
  def __init__(self,name,eyes,hands,legs,fins,breed,fur):
    super().__init__(name,eyes,hands,legs,fins)
    self.breed = breed
    self.fur = fur

  def bark(self):
    print('woof')
```

```
lokie = Dog('lokie',2,0,4,0,'St. Bernand','Brown')
```

```
lokie.eyes
```

```
lokie.fur
```

```
lokie.run()
```

```
lokie.swimming()
```

```
lokie.bark()
```
