```
class Sample:
  def __init__(self):
    pass

  def rock(self):
    raise NotImplementedError("Subclass must implement abstract method")
```

```
class Tester(Sample):
  pass
  def rock(self):
    print('rock')
```

```
test = Tester()
```

```
test.rock()
```

```
sample = Sample()
```

```
sample.rock()
```
