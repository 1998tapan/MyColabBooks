```
import re
```

```
patterns = ['term1', 'term2']
text = 'Is your term term1 or term2?'
```

```
# returns match object
re.search(patterns[0],text)
```

```
match = re.search(patterns[0], text)
```

```
# returns start index
match.start()
```

```
# returns end index
match.end()
```

```
# returns text
match.string
```

```
# returns length of text
match.endpos
```

```
# returns
match.group(0)
```

```
match.groups()
```

```
re.split(patterns[0],text)
```
