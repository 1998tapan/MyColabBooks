#Counter
Counter helps count hashable objects. It's a subclass of dict. It stores data in dict where object is key and it's count is value.

```
from collections import Counter
```

```
# counting occurences in list
# numbers
li = [1,11,2,3,3,2,2,1,1,1,1,11,11,2,2,2,3,3,3,35]
Counter(li)
```

```
print(Counter(li))
```

```
#let's try a list of characters
li = ['a','a','a','b','b','c','c','c','c','d','d','e']
Counter(li)
```

```
#using String
# counting occurences of letters
string = "aaabbbsdweififnifhfqdqwefwpgj"
Counter(string)
```

```
#but what if we want to find occurences of words
#easy - we split them to list !
string = "Hello Hello I am and you are are Hello Hello"
Counter(string.split())
```

#Counter methods

```
c = Counter(string)
```

```
c.most_common()
```

```
c.most_common(2)
```
