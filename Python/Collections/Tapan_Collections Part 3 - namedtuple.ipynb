#namedtuple

namedtuple is basically a normal tuple that defines a new object type and has name for each value.

Think of it as declaring a quick class.

```
from collections import namedtuple
Dog = namedtuple('Dog', 'name age breed')
```

```
sam = Dog('sammy', 12, 'labra')
```

```
type(sam)
```

```
sam[0]
```

```
sam
```

```
sam.name
```

```
sam.age
```
