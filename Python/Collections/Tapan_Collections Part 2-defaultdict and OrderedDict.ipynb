#defaultdict
Normal dict throws Key Error when a key doesn't exist. defaultdict doesn't. Instead it adds that key.

Using lambda, we can initialise it to 0


```
from collections import defaultdict
```

```
d = defaultdict(lambda:0)
```

```
d['one']
```

```
d['two'] = 2
```

```
for item in d:
  print(item)
```

```
d.items()
```

#OrderedDict
Ordered dictionaries

```
d = {}
d['a'] = 1
d['b'] = 2
d['c'] = 3
d['d'] = 4
d['e'] = 5
```

```
d
```

```
for k,v in d.items():
  print(k,v)
```

```
from collections import OrderedDict
d = OrderedDict()
d['a'] = 1
d['b'] = 2
d['c'] = 3
d['d'] = 4
d['e'] = 5
```

```
for k,v in d.items():
  print(k,v)
```

```
d
```

```
d1 = {1:1,2:2}
d2 = {2:2,1:1}

d1 == d2
```

```
d3 = OrderedDict(d1)
d4 = OrderedDict(d2)

d3 == d4
```
