#Generators
Generators fn allow us to write a fn that allows us to send back a value and then later resume to pick up where it left off.

Generators uses yield to return a current value

2 methods:

next() - gives next value in iterator and moves pointer by one

iter() - converts an obj to the iterable obj.



```
# fn without generators
def cube(n):
  result = []
  for i in range(n):
    result.append(i**3)
  return result
```

```
for num in cube(10):
  print(num)
```

```
# with generators
def gen_cubes(n):
  for i in range(n):
    yield i**3
```

```
cubes = gen_cubes(10)
```

```
print(cubes)
```

```
for num in cubes:
  print(num)
```

```
for num in gen_cubes(20):
  print(num)
```

```
# without generators
def fibon(n):
  result = []
  a = 0
  b = 1
  for _ in range(n):
    result.append(a)
    a,b = b,a+b
  
  return result
```

```
for num in fibon(10):
  print(num)
```

```
# with generators
def gen_fibon(n):
  a = 0
  b = 1
  for _ in range(n):
    yield a
    a,b = b,a+b
```

```
for num in gen_fibon(10):
  print(num)
```
