#Decorators

(Refer to Magical Bits of Python Part 1 to refresh some related concepts)

Decorators are used to add some new (created by the user) functionalities or features to fn to wrap them in to one single fn.

The good thing about decorators is you can enable/disable those extra features at your will.

Syntax:

@decorator

def some_fn():

fn body

Inorder to understand decorators and the background process done by @ operator, we will first create our own decorator

#Creating our own decorator

```
def need_decorator():
  print("I need a decorator")
```

```
def my_decorator(some_func):
  
  def wrap_fn():
    print("Code before the base fn")
    some_func()
    print("Code after the base fn")
  
  return wrap_fn
```

```
some_func = my_decorator(need_decorator)
```

```
some_func
```

```
some_func()
```

#using @ operator to use decorator

Just comment the @ line to remove extra functionalities

```
@my_decorator
def need_decorator():
  print("I need a decorator")
```

```
need_decorator()
```

```
#@my_decorator
def need_decorator():
  print("I need a decorator")
```

```
need_decorator()
```

# Use of decorators

Now, we as developers don't need to concern on creating our own decorators and use a wrap fn to wrap contents and return that fn. We just need to use decorators on our fn. But it's important to know how things work and the stuff that happens inside the @ operator.

Decorators are much used in web development to switch to a new design framework or library or something. It is extensively used in Django and Flask
