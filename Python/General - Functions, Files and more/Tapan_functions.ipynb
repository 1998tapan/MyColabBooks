```
word = "My DOG went for a walk"

print('Dog' in word.casefold())
```

```
def is_dog_in(word):
  return 'dog'in word.casefold()

print(is_dog_in("My DOG is sleeping."))
```

```
def pig_latin(word):
  #vowels = ['a','e','i','o','u'] 
  #both work
  vowels = 'aeiou'
  if word[0].casefold() in vowels:
    word+='ay'
  else:
    word = word[1:] + word[0] + 'ay'
  return word

print(pig_latin("Apple"))
```
