File methods()

read() - prints the file contents.

Note that, read() uses a virtual cursor that is initialised to beginning of file. Hence, after one read the cursor will be at the end of file and calling read() again would just print empty string.

To fix this, we need to reset cursor, this can be done using seek(pos_num)


```
myFile.read()
```

```
myFile.read()
```

```
myFile.seek(0)
```

```
myFile.read()
```

```
myFile.seek(0)
```
