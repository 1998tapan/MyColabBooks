```
timeit.timeit('"-".join(str(i) for i in range(100)) ', number=10000)
```

```
# using list comprehension
timeit.timeit('"-".join([str(i) for i in range(100)]) ', number = 10000)
```

```
#using map
timeit.timeit('"-".join(map(str,range(100))) ', number=10000)
```
